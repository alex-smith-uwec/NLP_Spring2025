# Import a corpus from [Project Gutenberg](https://www.gutenberg.org/)

```python
import requests

# Fetch book (Plain Text UTF-8)
url = "https://www.gutenberg.org/cache/epub/74409/pg74409.txt"
url = "https://www.gutenberg.org/cache/epub/70432/pg70432.txt"#Dorthy Sayers, Clouds of Witness
response = requests.get(url)
text = response.text

print(text[:500])
```

```python
# Strip headers and footers
start_index = text.find("*** START OF THE PROJECT GUTENBERG EBOOK CLOUDS OF WITNESS ***")
end_index = text.find("*** END OF THE PROJECT GUTENBERG EBOOK CLOUDS OF WITNESS ***")
corpus = text[start_index:end_index]

# Display first 500 characters
print(corpus[:500])
```

```python
print(corpus[800:850])
```

# Word Clouds

```python
import nltk
import numpy as np
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.draw.dispersion import dispersion_plot
from nltk.probability import FreqDist
from collections import Counter
from wordcloud import WordCloud
from matplotlib import pyplot as plt
```

```python
# Download NLTK resources
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
#Tokenize the corpus into words
words = word_tokenize(corpus.lower())

# Remove punctuation and stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

# Count word frequencies
word_counts = Counter(filtered_words)

# Display the most common words
print("Most common words:")
for word, count in word_counts.most_common(10):
    print(f"{word}: {count}")

# Perform additional analysis
total_words = len(filtered_words)
unique_words = len(word_counts)
lexical_diversity = unique_words / total_words

print("\nCorpus Analysis:")
print(f"Total words: {total_words}")
print(f"Unique words: {unique_words}")
print(f"Lexical diversity (unique/total): {lexical_diversity:.2f}")
```

```python
filtered_words[2000:2020]
```

```python
print(f"Number of words before removing stop words and non-alpha: {len(words)}")
print(f"Number of words after removing stop words and non-alpha (filtered_words)={len(filtered_words)}")
```

```python
all_fd=nltk.FreqDist(words)
filtered_fd=nltk.FreqDist(filtered_words)
```

```python
wordcloud_all = WordCloud(width = 1000, height = 800,
                      background_color ='white',
                      max_words=200,
                      prefer_horizontal=0.5,
                      # random_state=4,
                      contour_color='steelblue').generate_from_frequencies(all_fd)

wordcloud_filtered = WordCloud(width = 1000, height = 800,
                      background_color ='white',
                      max_words=200,
                      prefer_horizontal=0.5,
                      # random_state=4,
                      contour_color='steelblue').generate_from_frequencies(filtered_fd)
```

```python
# plot the WordCloud image: all words
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud_all)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.title("STUPID Word Cloud\n",fontsize=24)
plt.show()


```

```python
# plot the WordCloud image: filtered
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud_filtered)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.title("Smarter Word Cloud\n",fontsize=24)
plt.show()
```

# [Zipf's Law](https://en.wikipedia.org/wiki/Zipf%27s_law)

```python
words[100:120]
```

```python
cleaned_tokens=[word for word in words if word.isalpha()]
```

```python
cleaned_tokens[2000:2020]
```

```python
fdist = FreqDist(cleaned_tokens)
sorted_freqs = fdist.most_common()

sorted_freqs[0:10]
```

```python
word_frequencies = Counter(cleaned_tokens)
sorted_frequencies = sorted(word_frequencies.values(), reverse=True)

# Create an array of ranks (1, 2, 3, ...)
ranks = np.arange(1, len(sorted_freqs) + 1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ranks, sorted_frequencies, marker='.', linestyle='None')
plt.title("Zipf's Law Distribution")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Plot the data on a log-log scale
plt.figure(figsize=(10, 6))
plt.loglog(ranks, sorted_frequencies, marker='.', linestyle='None')
plt.title("Zipf's Law Distribution")
plt.xlabel("Rank (log scale)")
plt.ylabel("Frequency (log scale)")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import linregress

# Assuming you have already calculated ranks and sorted_frequencies as mentioned earlier
log_ranks = np.log(ranks)
log_frequencies = np.log(sorted_frequencies)

# Fit a linear regression model to the log-log data
slope, intercept, r_value, p_value, std_err = linregress(log_ranks, log_frequencies)

# The slope of the regression line represents the estimated exponent 's'
estimated_s = -slope

print(f"Estimated exponent 's' in Zipf's Law: {estimated_s}")
```
