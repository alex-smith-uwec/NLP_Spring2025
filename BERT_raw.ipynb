```python
import transformers
```

```python
from transformers import pipeline
```

```python
pipe = pipeline( model="google-bert/bert-base-uncased")
```

```python
from transformers import BertTokenizer, BertModel

# Initialize the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Encode some input text
inputs = tokenizer("Hello, my dog is cute", return_tensors="pt")

# Get model outputs
outputs = model(**inputs)

# The first token of every sequence is the [CLS] token
# Get its final hidden state
cls_hidden_state = outputs.last_hidden_state[:, 0, :]

print(f"shape of CLS token: {cls_hidden_state.shape}")
slice = cls_hidden_state[0][:5]

print(f"Some coordinates of CLS token: {slice}")
```

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")


prompt = "My [MASK] is Alice."
result = fill_mask(prompt)

# Display the results
for completion in result:
    print(completion)
```

```python
# Attempting to use BERT directly for sentiment analysis
# This won't work as intended because BERT is not fine-tuned for sentiment analysis

sentiment_analysis = pipeline("sentiment-analysis", model="bert-base-uncased")

sentiment_analysis("I love sunny days.")

```

```python
# Load a fine-tuned model for sentiment analysis
sentiment_analysis = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
sentiment_analysis("I love sunny days.")

```
