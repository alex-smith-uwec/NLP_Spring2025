```python
from transformers import GPT2Tokenizer

# Load the GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Get the vocabulary as a dictionary
vocab = tokenizer.get_vocab()

# Print the size of the vocabulary
print(f"Vocabulary size: {len(vocab)}")

# Display the first 20 tokens and their corresponding IDs
for i, (token, token_id) in enumerate(vocab.items()):
    if i < 20:
        print(f"Token: {token} \t ID: {token_id}")
    else:
        break
```

```python
vocab_items = list(vocab.items())
```

```python
vocab_items[3000:3050]
```

```python
from transformers import BertTokenizer

# Load multilingual BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

# Get the vocabulary
vocab = tokenizer.get_vocab()

# Print vocabulary size
print(f"Vocabulary size: {len(vocab)}")

# Display first  tokens
for i, (token, token_id) in enumerate(vocab.items()):
    if i < 10:
        print(f"Token: {token} \t ID: {token_id}")
    else:
        break
```
